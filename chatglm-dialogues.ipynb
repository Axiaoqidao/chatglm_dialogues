```python
def split_text_into_paragraphs(text):
    lines = text.split('\n')
    paragraphs, current_paragraph = [], ''

    for line in lines:
        if len(current_paragraph) + len(line) <= 300:
            current_paragraph += line + '\n'
        else:
            paragraphs.append(current_paragraph)
            current_paragraph = line + '\n'

    if current_paragraph:
        paragraphs.append(current_paragraph)

    merged_paragraphs, current_merged_paragraph = [], ''
    for paragraph in paragraphs:
        if len(current_merged_paragraph) + len(paragraph) <= 300:
            current_merged_paragraph += paragraph
        else:
            if len(current_merged_paragraph) >= 100:
                merged_paragraphs.append(current_merged_paragraph)
                current_merged_paragraph = paragraph
            else:
                current_merged_paragraph += paragraph

    if current_merged_paragraph:
        merged_paragraphs.append(current_merged_paragraph)

    return merged_paragraphs
```

```python
import csv

def 对话集(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        example_text = file.read()

    paragraphs = split_text_into_paragraphs(example_text)
    output = []

    for i, paragraph in enumerate(paragraphs):
        print(f'Paragraph {i + 1}:\n{paragraph}\n')
        response = process_paragraph(paragraph)
        if response:
            output.append((clean_text(paragraph), response))
            write_to_csv(output)

    return output

def process_paragraph(paragraph):
    max_attempts = 3
    string = []

    for attempt in range(max_attempts):
        try:
            response, history = model.chat(tokenizer, clean_text(paragraph), history=examples)
            string.append(response)
            print(response)
            print(eval(response))
            return eval(response)
        except Exception as e:
            print(e)
        finally:
            print(f"第 {attempt + 1} 次尝试完成。")

    if len(string) == 2:
        for attempt in range(max_attempts):
            try:
                response1, history1 = model.chat(tokenizer, string, history=edit)
                print(eval(response1))
                return eval(response1)
            except Exception as e:
                print(e)
            finally:
                print(f"修改完成。")

    return None

def clean_text(text):
    return text.replace('\u3000', '').replace('\n', '').replace(',', '，').replace("''", '’').replace('''"''', '“')

def write_to_csv(output):
    with open('text.csv', 'w', encoding='utf-8', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['content', 'summary'])
        writer.writerows(output)

print(对话集('/kaggle/input/guzhenren/guzhenren.txt'))
```
